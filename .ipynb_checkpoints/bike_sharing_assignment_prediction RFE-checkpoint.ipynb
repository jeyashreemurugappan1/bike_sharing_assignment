## Step 3: Data Preparation and conversion

- You can see that your dataset has many columns with values as '1' or '0'.We want to convert them to more meanigful format

```python
# convert season sit to meaninful categorical variable
varlist =  ['season']
def season_map(x):
    return x.map({1: "season_spring", 2: "season_summer",3:"season_fall",4:"season_winter"})
bike_sharing_data[varlist] = bike_sharing_data[varlist].apply(season_map)

varlist =  ['yr']
def year_map(x):
    return x.map({0:"year_2018", 1:"year_2019"})
bike_sharing_data[varlist] = bike_sharing_data[varlist].apply(year_map)

# convert weather sit to meaninful categorical variable
varlist =  ['weathersit']
# Defining the map function
def weather_map(x):
    return x.map({0:"weather_clear", 1:"weather_mist", 2:"weather_light_snow_rain",3:"weather_heavy_rain"})
bike_sharing_data[varlist] = bike_sharing_data[varlist].apply(weather_map)

# convert numerical variable date (1 to 31) to four bins D1, D2, D3, D4
bike_sharing_data['date_range'] = pd.cut(x=bike_sharing_data['date'], bins=[0,8,16,23,31],labels=['D1', 'D2', 'D3','D4'])

# convert month to meaninful categorical varaible
varlist =  ['mnth']
def month_map(x):
    return x.map({1:"jan", 2:"feb", 3:"mar",4:"apr",5:"may",6:"jun",7:"jul",8:"aug",9:"sep",10:"oct",11:"nov",12:"dec"})
bike_sharing_data[varlist] = bike_sharing_data[varlist].apply(month_map)

# convert weekday sit to meaninful categorical varaible. We arrive at this mapping by seeing the calendar
varlist =  ['weekday']
def weekday_map(x):
    return x.map({6:"monday", 0:"tuesday", 1:"wednesday",2:"thursday",3:"friday",4:"saturday",5:"sunday"})
bike_sharing_data[varlist] = bike_sharing_data[varlist].apply(weekday_map)
```

```python
bike_sharing_data.head()
```

### Dummy Variables

Handling dummy variables for below
- The variable `season` has four  levels 1:spring, 2:summer, 3:fall, 4:winter. 
- The variable `weather` has four levels clear , mist, light_snow_rain,heavy rain
- The variable `yr` has 2 levels 0: 2018 , 1:2019
- The variable `weekday` has 7 values 6:monday, 0:tuesday, 1:wednesday,2:thursday,3:friday,4:saturday,5:sunday
- The variable `mnth` has 12 values ranging from jan to dec
- The derived variable `date_range` has 4 values D1 : 1 to 8th, D2: 9th to 16th D3: 17th to 23rd,D4: 24th to 31st


```python
# handling dummies for season 
season_col = pd.get_dummies(bike_sharing_data['season'], drop_first = True)
bike_sharing_data = pd.concat([bike_sharing_data, season_col], axis = 1)
bike_sharing_data.drop(['season'], axis = 1, inplace = True)
#bike_sharing_data.drop(['season'], axis = 1, inplace = True)

#Handing dummies for 'weathersit' 
weathersit_col = pd.get_dummies(bike_sharing_data['weathersit'], drop_first = True)
bike_sharing_data = pd.concat([bike_sharing_data, weathersit_col], axis = 1)
bike_sharing_data.drop(['weathersit'], axis = 1, inplace = True)

#handling dummies for yr
yr_col = pd.get_dummies(bike_sharing_data['yr'], drop_first = True)
bike_sharing_data = pd.concat([bike_sharing_data, yr_col], axis = 1)
bike_sharing_data.drop(['yr'], axis = 1, inplace = True)

#handling dummies for weekday
weekday_col = pd.get_dummies(bike_sharing_data['weekday'], drop_first = True)
bike_sharing_data = pd.concat([bike_sharing_data, weekday_col], axis = 1)
bike_sharing_data.drop(['weekday'], axis = 1, inplace = True)

#handling dummies for month
month_col = pd.get_dummies(bike_sharing_data['mnth'], drop_first = True)
bike_sharing_data = pd.concat([bike_sharing_data, month_col], axis = 1)
bike_sharing_data.drop(['mnth'], axis = 1, inplace = True)

#handling dummies for date_range
date_col = pd.get_dummies(bike_sharing_data['date_range'], drop_first = True)
bike_sharing_data = pd.concat([bike_sharing_data, date_col], axis = 1)
bike_sharing_data.drop(['date_range'], axis = 1, inplace = True)

#As date range is there date is not required
bike_sharing_data.drop(['date'], axis = 1, inplace = True)
```

## Step 6 Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_pred = lm.predict(X_train_rfe)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

```python
mse_train = mean_squared_error(y_train, y_train_pred)
print('Mean_Squared_Error :' ,mse_train)
```

#### Mean squared error of 0.008908 means it indicates that model is not able to match 0.0089% of values only which is less than 0.1% which is good

```python
#checking r2 score of train
r_squared_train = r2_score(y_train, y_train_pred)
print("R square value",r_squared_train)
```

#### The r square value is about 82% which means our model is able to explain 82% of the variance which is also good
